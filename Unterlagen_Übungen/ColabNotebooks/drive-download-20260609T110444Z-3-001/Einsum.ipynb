# Einsum-Tutorial

Einsum ist eine Funktion in PyTorch/Numpy, die es uns ermöglicht, flexible und komplexe Operationen auf Tensoren durchzuführen. Einsum verwendet dafür eine kompakte Notation, die auf der Einstein-Summenkonvention basiert.
Um das Konzept besser zu verstehen, schauen wir uns einmal ein paar Beispiele an, für die wir einsum verwenden können:

1. Transponierte Matrix

```python
import torch
# Zunächst erzeugen wir einen 1D-Tensor "a" mithilfe von torch.arange(6). Unser Tensor a hat die Werte [0, 1, 2, 3, 4, 5]
# Mit .reshape formen wir unseren 1D-Tensor in eine Matrix (also einen 2D-Tensor) um.
a = torch.arange(6).reshape(2, 3)
print(f"Unser ursprünglicher Tensor a: \n {a}")

# Jetzt wollen wir einsum verwenden, um unsere Matrix zu transponieren.
# Alles, was wir dazu machen müssen, ist die Indizes des ursprünglichen Tensors zu vertauschen.
# In unserem urprünglichen Tensor haben wir zwei Dimensionen (2,3). Wir wählen entsprechend 2 Indizes i (für die Zeilen) und j (für die Spalten).
# Wir vertauschen die Indizes ij -> ji, haben jetzt also einen Tensor der Form (3,2)
# Note: in der eckigen Klammer steht eine Liste von Tensoren, auf die wir unsere Einsum-Operation anwenden (hier nur ein Tensor a)

b= torch.einsum('ij->ji', [a])
print(f"\nUnser neuer transponierter Tensor b: \n {b}")
```

2. Matrix-Vektor-Multiplikation

```python
a = torch.arange(6).reshape(3, 2)
print(f"Tensor a: \n {a}")
b = torch.arange(2)
print(f"\nVektor b: \n {b}")
# i und k respräsentieren hier die Zeilen (3) und Spalten (2) unserer Matrix a.
# k repräsentiert die Spalten in unserem Vektor (hier 2)
# und i die resultierenden Zeilen (hier eine)
c = torch.einsum('ik,k->i', [a, b])
print(f"\nMatrix-Vektor-Multiplikation mit einsum: \n {c}")

# selbes kann erreicht werden mit matmul:
d = torch.matmul(a, b)
print(f"\nMatrix-Vektor-Multiplikation mit matmul: \n {d}")

# Was hier konkret passiert:
# c[0] = 0*0 + 1*1 = 1
# c[1] = 2*0 + 3*1 = 3
# c[2] = 4*0 + 5*1 = 5
```

3. Matrix-Matrix-Multiplikation

```python
a = torch.arange(2).reshape(1, 2)
print(f"Tensor a: \n {a}")
b = torch.arange(3).reshape(1, 3)
print(f"\nTensor b: \n {b}")
print(torch.einsum("ik,ij->kj",a,b))
torch.einsum("ik,pj->kj",a,b)
```

```python
# Zunächst definieren wir auch hier zwei Matrizen
# Unsere Matrix a besitzt 6 Einträge, verteilt auf zwei Zeilen und drei Spalten
a = torch.arange(6).reshape(2, 3)
print(f"Tensor a: \n {a}")

# Unsere Matrix b besitzt 15 Einträge, verteilt auf drei Zeilen und fünf Spalten
b = torch.arange(15).reshape(3, 5)
print(f"\nTensor b: \n {b}")

# Beim Verwenden von einsum haben wir also pro Matrix zwei Indizes (weil unsere Matrizen 2D-Tensoren sind)
# Die Indizes unserer Matrix a heißen hier i und k, wobei i die Zeilen und k die Spalten repräsentiert
# Die Indizes unserer Matrix b heißen hier k und j, wobei k die Zeilen und j die Spalten repräsentiert
# Unser resultierender Tensor ist ebenfalls eine zweidimensionale Matrix mit 2 Zeilen und fünf Spalten.
# Der Tensor übernimmt also die Anzahl der Zeilen unserer Matrix a und die Anzahl der Spalten unserer Matrix b
c = torch.einsum('ik,kj->ij', [a, b])
print(f"\nMatrix-Matrix-Multiplikation mit einsum: \n {c}")

# selbes kann erreicht werden mit matmul:
d = torch.matmul(a, b)
print(f"\nMatrix-Matrix-Multiplikation mit matmul: \n {d}")
```

4. Skalarprodukt (Dot Product)

```python
# für Vektoren
a = torch.arange(5)
print(f"Vektor a: \n {a}")
# unser Vektor b soll Zahlen von 2 bis 6 beinhalten, d.h. er hat eine Länge von 5
# Es ist wichtig, dass unsere beiden Vektoren die gleiche Länge haben, damit wir das Skalarprodukt berechnen können
b = torch.arange(2,7)
print(f"\nVektor b: \n {b}")
c = torch.einsum('i,i->', [a, b])
print(f"\nSkalarprodukt: \n {c}")

# für Matrizen
a = torch.arange(6).reshape(2, 3)
print(f"\nMatrix a: \n {a}")
b = torch.arange(6,12).reshape(2, 3)
print(f"\nMatrix b: \n {b}")
c = torch.einsum('ij,ij->', [a, b])
print(f"\nSkalarprodukt: \n {c}")

# Berechnung:
# 1. jedes Element von a wird mit dem entsprechenden Element von b multipliziert
# 2. Ergebnisse werden summiert:

# Elementweise Multiplikation:
# (0 * 6) + (1 * 7) + (2 * 8) = 0 + 7 + 16 = 23 für die erste Zeile.
# (3 * 9) + (4 * 10) + (5 * 11) = 27 + 40 + 55 = 122 für die zweite Zeile.
# Summe der Ergebnisse: 23 + 122 = 145.
# Deshalb ist das Skalarprodukt c gleich 145.
```

5. Batch-Matrix-Multiplikation

```python
# Zunächst erzeugen wir zwei Tensoren, die jeweils 2-dimensional sein sollen und eine Batch-size von 3 haben sollen.
a = torch.randn(3,2,5)
print(f"Tensor a: \n {a}")
b = torch.randn(3,5,3)
print(f"\nTensor b: \n {b}")

# i steht für die Batch-size, j und k sind die Dimensionen des ersten Tensors (Matrix a) und k und l sind die Dimensionen des zweiten Tensors (Matrix b)
# unser resultierender Tensor hat ebenfalls die Batch-size i (hier 3) und die Dimensionen i und j (2,3), also zwei Zeilen und drei Spalten
c= torch.einsum('ijk,ikl->ijl', [a, b])
print(f"\nTensor nach Batch-Matrix-Multiplikation: \n {c}")

```
