# Linear Algebra


## 1. Scalars, Vectors, and Matrices

### Scalars
A **scalar** is just a single number:
- Examples: 5, -2.3, π
- In code: `x = 3.14`

### Vectors
A **vector** is an ordered list of numbers, typically written as a column:

```
v = [2]
    [1]
    [4]
```

**Key properties:**
- **Dimension**: Number of elements (this vector has dimension 3)
- **Components**: Individual elements (v₁ = 2, v₂ = 1, v₃ = 4)

**Geometric interpretation**: A vector represents a point in space or a direction from the origin.

### Matrices
A **matrix** is a rectangular array of numbers:

```
A = [1  2  3]
    [4  5  6]
```

**Key properties:**
- **Shape**: (rows × columns) - this matrix is 2×3
- **Elements**: A₍ᵢ,ⱼ₎ refers to row i, column j
- **Square matrix**: Same number of rows and columns

## 2. Vector Operations

### Vector Addition
Add corresponding components:
```
[2] + [1] = [3]
[3]   [4]   [7]
[1]   [2]   [3]
```

### Scalar Multiplication
Multiply each component by the scalar:
```
3 × [2] = [6]
    [1]   [3]
    [4]   [12]
```

### Dot Product (Inner Product)
Multiply corresponding components and sum:
```
[2]     [1]
[3]  ·  [4] = 2×1 + 3×4 + 1×2 = 2 + 12 + 2 = 16
[1]     [2]
```

**Geometric meaning**:
- dot(u, v) = ||u|| × ||v|| × cos(θ)
- Measures how "aligned" two vectors are
- If dot product = 0, vectors are perpendicular (orthogonal)

### Vector Length (Norm)
The **Euclidean norm** (L2 norm):
```
||v|| = √(v₁² + v₂² + ... + vₙ²)
```

For v = [3, 4]: ||v|| = √(3² + 4²) = √(9 + 16) = 5

## 3. Matrix Operations

### Matrix Addition
Add corresponding elements (matrices must have same shape):
```
[1  2] + [5  6] = [6   8]
[3  4]   [7  8]   [10  12]
```

### Matrix Multiplication
**Rule**: (m×n) matrix × (n×p) matrix = (m×p) matrix

The element at position (i,j) in the result is the dot product of row i from the first matrix and column j from the second matrix.

```
[1  2] × [5  6] = [1×5+2×7  1×6+2×8] = [19  22]
[3  4]   [7  8]   [3×5+4×7  3×6+4×8]   [43  50]
```

**Important**: Matrix multiplication is NOT commutative (AB ≠ BA in general)

### Matrix-Vector Multiplication
Treat the vector as a column matrix:
```
[1  2  3] × [1] = [1×1 + 2×2 + 3×3] = [14]
[4  5  6]   [2]   [4×1 + 5×2 + 6×3]   [32]
            [3]
```

**Interpretation**: Each row of the matrix takes a dot product with the vector.

### Transpose
The **transpose** of matrix A (written Aᵀ) flips rows and columns:
```
A = [1  2  3]    Aᵀ = [1  4]
    [4  5  6]         [2  5]
                      [3  6]
```

## 4. Special Matrices

### Identity Matrix
A square matrix with 1s on the diagonal and 0s elsewhere:
```
I₃ = [1  0  0]
     [0  1  0]
     [0  0  1]
```

**Property**: For any matrix A, AI = IA = A (like multiplying by 1)

### Zero Matrix
All elements are 0:
```
0 = [0  0  0]
    [0  0  0]
```

### Symmetric Matrix
A square matrix where A = Aᵀ:
```
[1  2  3]
[2  4  5]
[3  5  6]
```

## 5. Linear Combinations and Span

### Linear Combination
A **linear combination** of vectors v₁, v₂, ..., vₙ is:
```
c₁v₁ + c₂v₂ + ... + cₙvₙ
```

where c₁, c₂, ..., cₙ are scalars (coefficients).

**Example**:
```
2[1] + 3[0] = [2]
 [0]   [1]   [3]
```

### Span
The **span** of a set of vectors is all possible linear combinations of those vectors.

**Example**: The span of [1, 0] and [0, 1] is all of 2D space (ℝ²).

## 6. Linear Independence and Basis

### Linear Independence
Vectors are **linearly independent** if no vector can be written as a linear combination of the others.

**Test**: Vectors v₁, v₂, ..., vₙ are linearly independent if the only solution to:
```
c₁v₁ + c₂v₂ + ... + cₙvₙ = 0
```
is c₁ = c₂ = ... = cₙ = 0.

### Basis
A **basis** for a vector space is a set of linearly independent vectors that span the space.

**Standard basis for ℝ³**:
```
e₁ = [1]  e₂ = [0]  e₃ = [0]
     [0]       [1]       [0]
     [0]       [0]       [1]
```

## 7. Eigenvalues and Eigenvectors (Brief Introduction)

For a square matrix A, if Av = λv for some non-zero vector v and scalar λ, then:
- v is an **eigenvector** of A
- λ is the corresponding **eigenvalue**

**Intuition**: Eigenvectors are directions that don't change when the matrix transformation is applied (only scaled by λ).

## 8. Computational Linguistics Applications

### Word Vectors
In NLP, words are often represented as high-dimensional vectors:
```
"king" → [0.2, -0.1, 0.8, 0.4, ...]
"queen" → [0.3, -0.2, 0.7, 0.5, ...]
```

**Operations**:
- Similarity: dot product or cosine similarity
- Analogies: king - man + woman ≈ queen

### Document-Term Matrices
Represent documents and vocabulary as a matrix:
```
        "the"  "cat"  "sat"  "dog"
Doc1     3      2      1      0
Doc2     2      0      0      3
Doc3     1      1      1      1
```

### Neural Network Weights
Neural networks are essentially chains of matrix multiplications:
```
output = f(W₃ × f(W₂ × f(W₁ × input + b₁) + b₂) + b₃)
```

where Wᵢ are weight matrices, bᵢ are bias vectors, and f is an activation function.

### Attention Mechanisms
Attention in transformers uses matrix operations:
```
Attention(Q, K, V) = softmax(QKᵀ/√d)V
```

where Q (queries), K (keys), and V (values) are matrices derived from input embeddings.

## 9. Key Intuitions for NLP

### High-Dimensional Spaces
- Words live in high-dimensional vector spaces (often 100-1000 dimensions)
- Similar words are close together in this space
- Relationships between words correspond to directions in the space

### Matrix Transformations
- Neural networks learn to transform word vectors through matrix multiplications
- Each layer applies a linear transformation followed by a non-linear activation
- The goal is to map input representations to useful output representations

### Dimensionality
- **Input dimensionality**: Size of vocabulary or embedding dimension
- **Hidden dimensionality**: Size of intermediate representations
- **Output dimensionality**: Number of possible tags/classes (for supertagging)

## 10. Formulas Summary

**Vector dot product**: u · v = u₁v₁ + u₂v₂ + ... + uₙvₙ

**Vector norm**: ||v|| = √(v · v) = √(v₁² + v₂² + ... + vₙ²)

**Cosine similarity**: cos(θ) = (u · v) / (||u|| × ||v||)

**Matrix multiplication**: (AB)ᵢⱼ = Σₖ AᵢₖBₖⱼ

**Linear transformation**: y = Ax + b
